# **Minimum Spanning Tree (MST) Clustering**

MST clustering identifies clusters by looking at the minimum spanning tree of a distance graph.



```python
# Sample a subset of the data to reduce memory usage
df_sampled = df.sample(n=500, random_state=0)  # Adjust 'n' based on available memory
X = df_sampled.select_dtypes(include=[np.number]).fillna(0)
from sklearn.neighbors import NearestNeighbors
import networkx as nx
import matplotlib.pyplot as plt

# Use k-nearest neighbors to limit edges, reducing memory usage
k = 5  # Adjust k as needed
neighbors = NearestNeighbors(n_neighbors=k)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)

# Build MST graph with k-nearest neighbors
G = nx.Graph()
for i in range(len(X)):
    for j in range(1, k):  # Skip the first neighbor (the point itself)
        G.add_edge(i, indices[i][j], weight=distances[i][j])

# Compute MST from the k-nearest neighbors graph
mst = nx.minimum_spanning_tree(G)

# Visualize the MST
plt.figure(figsize=(10, 10))
nx.draw(mst, with_labels=False, node_size=50)
plt.title("Minimum Spanning Tree of Sampled Data")
plt.show()
```

# **Self-Organizing Maps (SOM)**



SOM is an unsupervised neural network for reducing dimensions, mapping high-dimensional data into a 2D grid.

```python
# Initialize and train the SOM
som = MiniSom(x=10, y=10, input_len=X.shape[1], sigma=0.5, learning_rate=0.5)
som.random_weights_init(X)
som.train_random(X, 100)  # Training for 100 iterations

# Plot SOM grid
plt.figure(figsize=(8, 8))
for i, x in enumerate(X):
    w = som.winner(x)  # Get the winning node
    plt.text(w[0] + 0.5, w[1] + 0.5, str(i), color=plt.cm.rainbow(i / len(X)), fontsize=10)

plt.title("Self-Organizing Map (SOM) Grid")
plt.grid()
plt.show()
```

# **Constructing a Linear Regression Model**

```python
# Convert target variable 'deposit' to numeric values
df['deposit'] = df['deposit'].map({'yes': 1, 'no': 0})

# Now redefine y
y = df['deposit']
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Prepare data (assuming df is already loaded)
X = df[['balance', 'age', 'duration']]  # Select relevant features
y = df['deposit']  # Target variable (adjust according to your dataset)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)
```

# **Evaluating the Model**

To evaluate a regression model, we use metrics such as Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared (R), each providing insight into the models accuracy.

```python
# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)  # Root Mean Squared Error
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R):", r2)
```
