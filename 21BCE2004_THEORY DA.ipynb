## **Importing required libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Load the dataset**

```python
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bank.csv"
df = pd.read_csv(url)
```

**Displaying the first few rows**

```python
print("First few rows of the dataset:")
print(df.head())
```

```python
print("\nShape of the dataset:", df.shape)
print("\nDataset information:")
df.info()
```

```python
# Check basic statistics
df.describe()
```

**Checking Data Types**


```python
# Checking the data types
print("Data types of columns:\n", df.dtypes)
```

**Checking for missing values**

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Dropping rows with any missing values
df_dropped = df.dropna()
print(df_dropped)
```

## **Filling Missing values**

```python
# Fill missing values with 0
df_filled = df.fillna(0)
print(df_filled)
```

```python
# Forward fill
df_ffill = df.fillna(method='ffill')
print(df_ffill)
```

```python
# Backward fill
df_bfill = df.fillna(method='bfill')
print(df_bfill)
```

## **Replacing Values**


```python
# Replace all occurrences of 'unknown' in the dataset with NaN
df_replaced = df.replace('unknown', np.nan)
print(df_replaced)
```

```python
# Checking for duplicates
print(df.duplicated())
```

```python
# Dropping duplicates
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates)
```

# **Discretization and Binning**


```python
# Binning the 'age' column into intervals
age_bins = [18, 30, 40, 50, 60, 100]
age_labels = ['18-30', '31-40', '41-50', '51-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
print(df['age_group'].value_counts())
```

# **Outlier Detection and Filtering**

```python
# Detecting outliers in 'balance' column using a threshold
threshold = df['balance'].mean() + 3 * df['balance'].std()
outliers = df[df['balance'] > threshold]
print(outliers)
```

# **Permutation and Random Sampling**

```python
# Permutation of rows
permuted_df = df.take(np.random.permutation(len(df))[:10])
print(permuted_df)
```

# **Merging and Concatenating DataFrames**

## **Concatenating along an axis**

```python
# Creating sample DataFrames
df1 = df.iloc[:5, :5]  # First 5 rows and 5 columns
df2 = df.iloc[5:10, :5]
```

```python
# Concatenate along rows (axis=0)
concat_df = pd.concat([df1, df2], ignore_index=True)
print(concat_df)
```

```python
# Concatenate along columns (axis=1)
concat_df_axis1 = pd.concat([df1, df2], axis=1)
print(concat_df_axis1)
```

## **Using merge with different types of joins**

```python
# Creating two DataFrames with a common column 'age'
df3 = df[['age', 'balance', 'job']].iloc[:5]
df4 = df[['age', 'marital', 'education']].iloc[3:8]
```

```python
# Inner Join
inner_merge = df3.merge(df4, on='age', how='inner')
print(inner_merge)
```

```python
# Left Join
left_merge = df3.merge(df4, on='age', how='left')
print(left_merge)
```

```python
# Right Join
right_merge = df3.merge(df4, on='age', how='right')
print(right_merge)
```

```python
# Outer Join
outer_merge = df3.merge(df4, on='age', how='outer')
print(outer_merge)
```

# **Time Series Analysis**

```python
print("\nTime Series Analysis:")
```

```python
# Creating a time series index
df['date'] = pd.date_range(start='1/1/2020', periods=len(df), freq='D')
df.set_index('date', inplace=True)
df.head()
```

```python
# Plotting the 'balance' over time
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['balance'], color='blue')
plt.title('Time Series Plot of Balance')
plt.xlabel('Date')
plt.ylabel('Balance')
plt.show()
```

```python
# Resampling the data to monthly frequency and calculating the mean
monthly_balance = df['balance'].resample('M').mean()
```

```python
# Plot the resampled data
plt.figure(figsize=(10, 5))
plt.plot(monthly_balance, marker='o', color='red')
plt.title('Monthly Average Balance')
plt.xlabel('Month')
plt.ylabel('Average Balance')
plt.show()
```

# **Rolling statistics**

```python
# Calculating rolling mean and standard deviation
rolling_mean = df['balance'].rolling(window=30).mean()
rolling_std = df['balance'].rolling(window=30).std()
```

```python
# Plotting rolling statistics
plt.figure(figsize=(10, 5))
plt.plot(df['balance'], label='Original', color='blue')
plt.plot(rolling_mean, color='red', label='Rolling Mean')
plt.plot(rolling_std, color='black', label='Rolling Std Dev')
plt.title('Rolling Mean & Standard Deviation')
plt.legend()
plt.show()
```
